```python
def next_words(token,seq,input_word,num_word):
    in_text=input_word
    result=input_word
    for _ in range(num_word):
        encod=token.texts_to_sequences([in_text])[0]
        encoding=array(encod)
        y_pred=seq.predict_classes(encoding)
        word_out=''
        for word,index in token.word_index.items():
            if index == y_pred:
                word_out=word
                in_text = word_out
                result+=' ' + word_out
                break
        
      
    return result
```

```python
token=Tokenizer()
token.fit_on_texts([dat])
v_size=len(token.word_index)+1
seq=list()
for line in dat.split('\n'):
    end=token.texts_to_sequences([line])[0]
    for i in range(1,len(end)):
        encod=end[:i+1]
        seq.append(encod)
```

```python
print(len(seq))
```

```python
max_length = max([len(se) for se in seq])
seq=keras.preprocessing.sequence.pad_sequences(seq,maxlen=max_length,padding='pre')
```

```python
seq=array(seq)
```

```python
x=seq[:,:-1]
y=seq[:,-1]
y = to_categorical(y, num_classes=v_size)
```

```python
seqq=Sequential()
seqq.add(Embedding(v_size,10,input_length=max_length-1))
seqq.add(LSTM(50))
seqq.add(Dense(units=v_size,activation='softmax'))
seqq.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
seqq.fit(x,y,epochs=500,verbose=2)
```

```python
def next_words(token,seq,max_length,input_word,num_word):
    in_text=input_word
    
    for _ in range(num_word):
        encod=token.texts_to_sequences([in_text])[0]
        encoding = keras.preprocessing.sequence.pad_sequences([encod], maxlen=max_length, padding='pre')
        y_pred=seq.predict_classes(encoding)
        word_out=''
        for word,index in token.word_index.items():
            if index == y_pred:
                
                in_text +=' '+ word
                
                break
        
      
    return in_text
```
